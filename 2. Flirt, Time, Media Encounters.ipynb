```python
# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%y')

# Strip extra spaces and special characters from 'Time' column
df['Time'] = df['Time'].str.strip()

# Convert 'Time' column to datetime
df['Time'] = pd.to_datetime(df['Time'], format='%I:%M %p').dt.time

# Combine 'Date' and 'Time' columns into a single datetime column
df['DateTime'] = pd.to_datetime(df['Date'].astype(str) + ' ' + df['Time'].astype(str))

# Extract hour from the datetime
df['Hour'] = df['DateTime'].dt.hour

most_active_hour = df.groupby(['Date', 'Hour']).size().reset_index(name='Count')
most_active_hour = most_active_hour.loc[most_active_hour.groupby('Date')['Count'].idxmax()]
```
